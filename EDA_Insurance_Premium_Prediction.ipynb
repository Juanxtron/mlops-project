```python
# Importar las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilos de visualización
sns.set(style="whitegrid")
%matplotlib inline

# Cargar los datasets
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

# Mostrar las primeras filas de ambos datasets
print("Train Dataset:")
display(train_df.head())

print("\nTest Dataset:")
display(test_df.head())

# Información general de los datasets
print("Información del Train Dataset:")
train_df.info()

print("\nInformación del Test Dataset:")
test_df.info()

# Descripción estadística de variables numéricas
print("Descripción Estadística del Train Dataset:")
display(train_df.describe())

print("\nDescripción Estadística del Test Dataset:")
display(test_df.describe())
```

```python
# Análisis de valores faltantes
def missing_data_analysis(df, dataset_name):
    missing_values = df.isnull().sum()
    missing_percentage = (missing_values / len(df)) * 100
    missing_df = pd.DataFrame({
        'Column': df.columns,
        'Missing Values': missing_values,
        'Missing Percentage': missing_percentage
    })
    missing_df = missing_df[missing_df['Missing Values'] > 0]
    missing_df = missing_df.sort_values(by='Missing Percentage', ascending=False)
    print(f"\nValores Faltantes en {dataset_name}:")
    display(missing_df)

missing_data_analysis(train_df, "Train Dataset")
missing_data_analysis(test_df, "Test Dataset")
```

```python
# Análisis de tipos de datos incorrectos
def datatype_analysis(df, dataset_name):
    print(f"\nTipos de Datos en {dataset_name}:")
    display(df.dtypes)

datatype_analysis(train_df, "Train Dataset")
datatype_analysis(test_df, "Test Dataset")
```

```python
# Análisis de distribuciones de variables numéricas
def distribution_analysis(df, dataset_name):
    num_cols = df.select_dtypes(include=['int64', 'float64']).columns
    for col in num_cols:
        plt.figure(figsize=(8, 4))
        sns.histplot(df[col].dropna(), kde=True)
        plt.title(f'Distribución de {col} en {dataset_name}')
        plt.show()


distribution_analysis(train_df, "Train Dataset")
distribution_analysis(test_df, "Test Dataset")
```

```python
# Análisis de variables categóricas
def categorical_analysis(df, dataset_name):
    cat_cols = df.select_dtypes(include=['object']).columns
    for col in cat_cols:
        plt.figure(figsize=(8, 4))
        sns.countplot(data=df, x=col)
        plt.title(f'Conteo de {col} en {dataset_name}')
        plt.xticks(rotation=45)
        plt.show()

categorical_analysis(train_df, "Train Dataset")
categorical_analysis(test_df, "Test Dataset")
```

```python
# Identificación de outliers
def outlier_analysis(df, dataset_name):
    num_cols = df.select_dtypes(include=['int64', 'float64']).columns
    for col in num_cols:
        plt.figure(figsize=(8, 4))
        sns.boxplot(x=df[col])
        plt.title(f'Boxplot de {col} en {dataset_name}')
        plt.show()

outlier_analysis(train_df, "Train Dataset")
outlier_analysis(test_df, "Test Dataset")
```

```python
# Análisis de correlaciones
def correlation_analysis(df, dataset_name):
    plt.figure(figsize=(12, 10))
    corr = df.corr()
    sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
    plt.title(f'Matriz de Correlación en {dataset_name}')
    plt.show()

correlation_analysis(train_df, "Train Dataset")
correlation_analysis(test_df, "Test Dataset")
```
